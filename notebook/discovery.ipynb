```python
result = sqldf("""
SELECT
    customer.id, SUM(sales.price_sold_per_device) AS revenue_by_customer FROM df_sales sales 
    LEFT JOIN df_country country ON sales.country_id=country.id
    LEFT JOIN df_customer customer ON sales.customer_id=customer.id GROUP BY customer.id 
""", locals())
```
